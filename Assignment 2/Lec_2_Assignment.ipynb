```python
### Variables 
calculationType = ""
shapeType = ""
PI = 3.14

### Get Calculation Type (1 is Circumference) and (2 is Area)
while True:
    print("please Choose Type of Calculation : ")
    calculationType = input("Press [1] to choose Circumference\nPress [2] to choose Area\nYour Choice is: ")
    if calculationType == "1" or calculationType == "2":
        break;
    else:
        print("Calculation Type Choice Is Invalid")
        
### Get Shape Type (1 is Circule), (2 is Rectangle) and (2 is Square)
while True:
    shapeType = input("Press [1] to choose Circule\nPress [2] to choose Rectangle\nPress [3] to choose square\nYour Choice is: ")
    if shapeType == "1" or shapeType == "2" or shapeType == "3":
        break;
    else:
        print("Shape Type Choice Is Invalid")

#########################################
############## User Choice ##############
choosenCalculation = ""
if calculationType=="1":
    choosenCalculation = "Circumference"
elif calculationType=="2":
    choosenCalculation="Area"

choosenShape = ""
if shapeType=="1":
    choosenShape = "Circule"
elif shapeType=="2":
    choosenShape="Rectangle"
elif shapeType=="3":
    choosenShape="square"
print("***************************************************************************")
print("***************************************************************************")
print("Your Choice is : Calculate The ",choosenCalculation," Of ",choosenShape)
print("***************************************************************************")
print("***************************************************************************")
    
################ Circule ################         
if shapeType=="1":
    radius = float(input("Please Enter Circule Radius : "))
    ### Circumference
    if calculationType == "1":
        Circumference = 2 * PI * radius
        print(" - Circule Circumference is = ",Circumference)
    ### Area
    elif  calculationType == "2":
        Area = PI * radius * radius
        print(" - Circule Area is = ",Area)
        
############### Rectangle ###############
elif  shapeType=="2":
    Height = float(input("Please Enter Rectangle Height : "))
    Width = float(input("Please Enter Rectangle Width : "))
    ### Circumference
    if calculationType == "1": 
        Circumference = (2 * Height) + (2 * Width)
        print(" - Rectangle Circumference is = ",Circumference)
    ### Area
    elif  calculationType == "2":
        Area = Height * Width
        print(" - Rectangle Area is = ",Area)   
    
############### Square ###############
elif  shapeType=="3":
    Side = float(input("Please Enter Square Side : "))
    ### Circumference
    if calculationType == "1": 
        Circumference = (4 * Side)
        print(" - Square Circumference is = ",Circumference)
    ### Area
    elif  calculationType == "2": 
        Area = (Side * Side)
        print(" - Square Area is = ",Area)

print("\n\n\n                 - Thanks! You Can Fine Me At...")
print("                         email : eng.mahmoud.ahmed.mohamed@gmail.com")
```
